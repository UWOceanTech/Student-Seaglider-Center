```python
#!pip install netCDF4
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from netCDF4 import Dataset
from datetime import datetime
import xarray as xr
from matplotlib.ticker import ScalarFormatter
import cmocean as cm
```

```python
os.chdir(r"C:/Users/lydia/Seagliders/tropics_mission/tropics_data") # sets my working directory
path = 'dives_2025'
files = sorted(os.listdir(path))
```

```python
# Initialize lists to store combined data
salinity = []
density = []
depth = []
temp = []
time = []
lat = []
lon = []
oxy = []
```

```python
# Loop through files and collect data
# I have the variables that I am not using commented out to save time when running the code
for file in files:
    dataset = xr.open_dataset(os.path.join(path, file))
    # Extract variables
    #salinity_values = dataset['salinity'].values.flatten()
    density_values = dataset['density'].values.flatten()
    depth_values = dataset['depth'].values.flatten()
    #temp_values = dataset['temperature'].values.flatten()
    #time_values = dataset['time'].values.flatten()
    lat_values = dataset['latitude'].values.flatten()
    lon_values = dataset['longitude'].values.flatten()
    #oxy_values = dataset['dissolved_oxygen_sat'].values.flatten()
    
    # Append to lists
    #salinity.extend(salinity_values)
    density.extend(density_values)
    depth.extend(depth_values)
    #temp.extend(temp_values)
    #time.extend(time_values)
    lat.extend(lat_values)
    lon.extend(lon_values)
    #oxy.extend(oxy_values)

# Convert lists to NumPy arrays for plotting
#salinity = np.array(salinity)
density = np.array(density)
depth = np.array(depth)
#temp = np.array(temp)
#time = np.array(time)
lat = np.array(lat)
lon = np.array(lon)
#oxy = np.array(oxy)
```

```python
# Plot everything on one figure
plt.figure(figsize=(15, 6))
scatter = plt.scatter(lon, lat, c="darkcyan", s=25)
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Seaglider Track in the Tropical Pacific")

plt.figure(figsize=(15, 6))
scatter = plt.scatter(lon, depth, c=density, cmap= cm.cm.dense, s=80)
plt.colorbar(label="Density")
plt.xlabel("Longitude")
plt.ylabel("Depth")
plt.gca().invert_yaxis()

plt.title("Density Variation across Seaglider Track (2025)")

plt.show()
```

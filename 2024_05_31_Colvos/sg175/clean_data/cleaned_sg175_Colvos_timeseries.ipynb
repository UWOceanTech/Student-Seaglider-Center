```python
import xarray as xr
import pandas as pd
import os
```

```python
print(os.getcwd())
#Load the dataset
colvos_ds = xr.open_dataset('sg175_UWSSC_Colvos_Mission_timeseries.nc')
display(colvos_ds)
# check time and depth values
colvos_ds.depth.values
colvos_ds.time.values
```

```python
#Align time with sg_data_point and apply offset
adjusted_time = pd.to_datetime(colvos_ds['time'].values) + pd.DateOffset(years=0, months=0, days=0)

colvos_ds = colvos_ds.assign_coords(time=('sg_data_point', adjusted_time))

#Select the relevant variables
new_colvos_ds = colvos_ds[['time', 'depth', 'latitude', 'longitude','temperature', 'salinity']] 
print(new_colvos_ds['time'][0])

#Convert to DataFrame and save
new_colvos_ds.to_dataframe().reset_index().to_csv('sg175_Colvos_Mission_timeseries_cleaned.csv', index=False)
new_colvos_ds.to_netcdf('sg175_Colvos_Mission_timeseries_cleaned.nc')
```

```python
colvos_ds = xr.open_dataset('sg175_UWSSC_Colvos_Mission_timeseries.nc')

#Apply time apply offset (if needed)
adjusted_start = pd.to_datetime(colvos_ds['start_time'].values) + pd.DateOffset(years=0, months=0, days=0)
adjusted_end = pd.to_datetime(colvos_ds['end_time'].values) + pd.DateOffset(years=0, months=0, days=0)
colvos_ds = colvos_ds.assign_coords(start_time=('dive', adjusted_start), end_time=('dive', adjusted_end))

colvos_ds['U_DAC'] = colvos_ds['depth_avg_curr_east']
colvos_ds['V_DAC'] = colvos_ds['depth_avg_curr_north']

# add metadata
colvos_ds['U_DAC'].attrs['pre_cleaning_name'] = 'depth_avg_curr_east'
colvos_ds['V_DAC'].attrs['pre_cleaning_name'] = 'depth_avg_curr_north'

#Select the relevant variables
new_colvos_ds = colvos_ds[['U_DAC', 'V_DAC', 'start_time', 'end_time', 'start_latitude', 'end_latitude', 'start_longitude', 'end_longitude']]
display(new_colvos_ds)

#Convert to DataFrame and save
new_colvos_ds.to_dataframe().reset_index().to_csv('sg175_Colvos_Mission_timeseries_DAC_cleaned.csv', index=False)
new_colvos_ds.to_netcdf('sg175_Colvos_Mission_timeseries_DAC_cleaned.nc')
```
